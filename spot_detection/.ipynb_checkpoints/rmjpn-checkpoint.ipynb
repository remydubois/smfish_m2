```python
%matplotlib inline
from skimage.morphology import selem
import numpy
import matplotlib.pyplot as plt

def _ellipse_in_shape(shape, center, radii):
    """Generate coordinates of points within ellipse bounded by shape.
    Parameters
    ----------
    shape :  iterable of ints
        Shape of the input image.  Must be length 3.
    center : iterable of floats
        (row, column) position of center inside the given shape.
    radii : iterable of floats
        Size of two half axes (for row and column)
    rotation : float, optional
        Rotation of the ellipse defined by the above, in radians
        in range (-PI, PI), in contra clockwise direction,
        with respect to the column-axis.
    Returns
    -------
    rows : iterable of ints
        Row coordinates representing values within the ellipse.
    cols : iterable of ints
        Corresponding column coordinates representing values within the ellipse.
    """
    r_lim, c_lim, d_lim = numpy.ogrid[0:float(shape[0]), 0:float(shape[1]), 0:float(shape[2])]
    r_org, c_org, d_org = center
    r_rad, c_rad, d_rad = radii
    r, c, d = (r_lim - r_org), (c_lim - c_org), (d_lim - d_org)
    distances = (r / r_rad) ** 2 \
                + (c / c_rad) ** 2 \
                + (d / d_rad) ** 2
    return numpy.nonzero(distances < 1)

def ellipsis(radii):
    if len(radii) < 3:
        raise ValueError('Please use skimage ellipse for 2d ellipse. Otherwise provide three radii.')
    
    selem = numpy.zeros(tuple(map(lambda x: 2*x + 1, radii)))
    
    mask = _ellipse_in_shape(selem.shape, tuple(map(lambda x: x, radii)), radii)
    
    selem[mask] = 1
    
    return selem

m = ellipsis((10, 12, 10))
f,ax = plt.subplots()
ax.imshow(m[:, :, 5])
f.show()
```

```python
import functools
import numpy
from image import *
from skimage.util import img_as_ubyte
from skimage.morphology import  ball
from skimage.filters.rank import generic_cy
from scipy import ndimage as ndi


def _handle_inumpyut(image, selem, out, mask, out_dtype=None, pixel_size=1):

    # assert_nD(image, 2)
    if image.dtype not in (numpy.uint8, numpy.uint16):
        image = img_as_ubyte(image)

    selem = numpy.ascontiguousarray(img_as_ubyte(selem > 0))
    image = numpy.ascontiguousarray(image)

    if mask is None:
        mask = numpy.ones(image.shape, dtype=numpy.uint8)
    else:
        mask = img_as_ubyte(mask)
        mask = numpy.ascontiguousarray(mask)

    if image is out:
        raise NotImplementedError("Cannot perform rank operation in place.")

    if out is None:
        if out_dtype is None:
            out_dtype = image.dtype
        out = numpy.empty(image.shape+(pixel_size,), dtype=out_dtype)
    else:
        if len(out.shape) == 2:
            out = out.reshape(out.shape+(pixel_size,))

    is_8bit = image.dtype in (numpy.uint8, numpy.int8)

    if is_8bit:
        max_bin = 255
    else:
        max_bin = max(4, image.max())

    bitdepth = int(numpy.log2(max_bin))
    # if bitdepth > 10:
    #     warn("Bitdepth of %d may result in bad rank filter "
    #          "performance due to large number of bins." % bitdepth)

    return image, selem, out, mask, max_bin


def _apply_scalar_per_pixel(func, image, selem, out, mask, shift_x, shift_y,
                            out_dtype=None):

    image, selem, out, mask, max_bin = _handle_inumpyut(image, selem, out, mask,
                                                     out_dtype)

    func(image, selem, shift_x=shift_x, shift_y=shift_y, mask=mask,
         out=out, max_bin=max_bin)

    return out.reshape(out.shape[:image.ndim])

def mean(image, selem, out=None, mask=None, shift_x=False, shift_y=False):
    """Return local mean of an image.
    Parameters
    ----------
    image : 2-D array (uint8, uint16)
        Inumpyut image.
    selem : 2-D array
        The neighborhood expressed as a 2-D array of 1's and 0's.
    out : 2-D array (same dtype as inumpyut)
        If None, a new array is allocated.
    mask : ndarray
        Mask array that defines (>0) area of the image included in the local
        neighborhood. If None, the complete image is used (default).
    shift_x, shift_y : int
        Offset added to the structuring element center point. Shift is bounded
        to the structuring element sizes (center must be inside the given
        structuring element).
    Returns
    -------
    out : 2-D array (same dtype as inumpyut image)
        Output image.
    Examples
    --------
    >>> from skimage import data
    >>> from skimage.morphology import disk
    >>> from skimage.filters.rank import mean
    >>> img = data.camera()
    >>> avg = mean(img, disk(5))
    """

    return _apply_scalar_per_pixel(generic_cy._mean, image, selem, out=out,
                                   mask=mask, shift_x=shift_x, shift_y=shift_y)

im = FQimage()
im.load(get_image_path())
 
mean(im.image_raw, ball(4))
```
